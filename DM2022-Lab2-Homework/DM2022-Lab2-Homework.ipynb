## Feature engineering

Use bag of word and nltk.word_tokenize to create 500 feature for each text.

8 output categories are represented by one-hot coding


## Model

Input shape = 500

3 hidden layer:
1st : 500->256 -> RELU()
2nd : 256->128 -> RELU()
3rd : 128->64 -> RELU()

Output layer: 64->8 -> softmax()

All layers are fully connected

optimizer : adam
loss : categorical_crossenttropy
epoch = 25
batch_size = 128
