```python
import pandas as pd
identification = pd.read_csv("data/data_identification.csv")
emotion = pd.read_csv("data/emotion.csv")
```

```python
train = identification[identification.identification=="train"]
test = identification[identification.identification=="test"]
```

```python
import json
data_list = []
with open("data/tweets_DM.json", "r") as f:
    for line in f:
        data_list.append(json.loads(line)["_source"]["tweet"])
```

```python
data_list
```

```python
tweets = pd.DataFrame(data_list)
```

```python
tweets
```

```python
train_data = pd.merge(train, tweets, on = "tweet_id", how="left")
train_data = pd.merge(train_data, emotion, on = "tweet_id")
train_data
```

```python
test_data = pd.merge(test, tweets, on = "tweet_id", how = "left")
test_data
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
# build analyzers (bag-of-words)
BOW_500 = CountVectorizer(max_features=500, tokenizer=nltk.word_tokenize) 

# apply analyzer to training data
BOW_500.fit(train_data['text'])
```

```python
import keras

# standardize name (X, y) 
X_train = BOW_500.transform(train_data['text'])
y_train = train_data['emotion']

X_test = BOW_500.transform(test_data['text'])


## check dimension is a good habbit 
print('X_train.shape: ', X_train.shape)
print('y_train.shape: ', y_train.shape)
print('X_test.shape: ', X_test.shape)
```

```python
input_shape = X_train.shape[1]
print('input_shape: ', input_shape)

output_shape = len(label_encoder.classes_)
print('output_shape: ', output_shape)
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers import ReLU, Softmax

# input layer
model_input = Input(shape=(input_shape, ))  # 500
X = model_input

# 1st hidden layer
X_W1 = Dense(units=256)(X)  # 64
H1 = ReLU()(X_W1)

# 2nd hidden layer
H1_W2 = Dense(units=128)(H1)  # 64
H2 = ReLU()(H1_W2)


H2_W3 = Dense(units=64)(H2)  # 64
H3 = ReLU()(H2_W3)
# output layer
H3_W4 = Dense(units=output_shape)(H3)  # 8
H4 = Softmax()(H3_W4)

model_output = H4

# create model
model = Model(inputs=[model_input], outputs=[model_output])

# loss function & optimizer
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# show model construction
model.summary()
```

```python
from keras.callbacks import CSVLogger



# training setting
epochs = 25
batch_size = 128

# training!
history = model.fit(X_train, y_train, 
                    epochs=epochs, 
                    batch_size=batch_size)
print('training finish')
```
